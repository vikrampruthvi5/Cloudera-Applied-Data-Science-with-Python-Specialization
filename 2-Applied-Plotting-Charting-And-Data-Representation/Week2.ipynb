Scatter Plot

```python
x = [1, 2, 3 ,4, 5]
y = x

colors = ['orange', 'blue', 'green' , 'black', 'red']

plt.figure()

plt.scatter(x[:2], y[:2], c=['blue'], s= 100, label='Short')
plt.scatter(x[2:], y[2:], c=['red'], s= 100, label='Tall')

plt.xlabel('x label info')
plt.ylabel('y label info')
plt.title('Sample scatter plot')
plt.legend()
plt.legend(loc=4, title='legend', frameon=False)
```

Line Plots

```python
import numpy as np

linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8])
quad_data = linear_data**2

plt.figure()

# creating 2 lines
plt.plot(linear_data, 'r-o', quad_data, 'g-o')

#adding more lines
n_dat1 = [22, 44, 55]
plt.plot(n_dat1, '--r')

# adding properties
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.legend(['US', 'India', 'China'], frameon=False, title='Country')

# fill between
plt.gca().fill_between(range(len(linear_data)), linear_data, quad_data, facecolor='orange', alpha=0.25)
plt.gca().fill_between(range(len(n_dat1)), n_dat1, quad_data[:len(n_dat1)], facecolor='purple', alpha=0.25)
plt.gca().fill_between(range(len(n_dat1)), linear_data[:len(n_dat1)], quad_data[:len(n_dat1)], facecolor='red', alpha=0.25)

# Rotating axis information
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
    
# Adjust borders
plt.subplots_adjust(bottom=0.15)
```

```python
langs = ['python', 'sql', 'java', 'cpp', 'Java Script']
pops = [56, 39, 34, 34, 29]

plt.figure()
plt.bar(langs[:1], pops[:1], width=0.5, align='center')
plt.bar(langs[1:], pops[1:], width=0.5, align='center')

plt.xlabel('languages')
plt.ylabel('popularity')
plt.title('Languages and their popularities')

# remove ticks on both axes
plt.tick_params(bottom=False, left=False, right=False, top=False, labelbottom=True)

# remove border for plot
for i in plt.gca().spines.values():
   i.set_visible(False)
    
# All y labels turned to grey
for i in plt.gca().xaxis.get_ticklabels():
   i.set_color('grey')


#plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = plt.gcf()
y = plt.gca()
x.clear()
y.clear()

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change the bar colors to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')
# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```

```python
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [56, 39, 34, 34, 29]

import matplotlib.pyplot as plt

# Creating bar graph with minimalistic data-ink
plt.figure()

# Remove tick marks
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# Clear borders using spines
for spine in plt.gca().spines.values():
    spine.set_visible(False)
    
bars = plt.bar(languages, popularity, alpha=0.6)
bars[0].set_color('grey')
```
